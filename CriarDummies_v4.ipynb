```python
# Função para verificar concorrência de processos com 100% do tempo
def verificarConcorrencia_Processo(processo, dtIni, dtFim):
    # iterar em todas execuções do referido processo
    for i in df[df["Task"]==processo].sort_values(by=["Start"], ascending=True).index: 
        # a última parte do OR é para pegar os casos de 100% de concorrência
        if (df[df["Task"]==processo]["Start"][i]>= dtIni and df[df["Task"]==processo]["Start"][i]<= dtFim) or (df[df["Task"]==processo]["Finish"][i]>= dtIni and df[df["Task"]==processo]["Finish"][i]<= dtFim) or (df[df["Task"]==processo]["Start"][i]<= dtIni and df[df["Task"]==processo]["Finish"][i]>= dtFim):
            # pega início e fim do processo concorrente
            StartConcorrente = df["Start"][i]
            FinishConcorrente = df["Finish"][i]
            # adequa início e fim do processo concorrente para dentro do range do processo principal para estimar tempo rodando em paralelo
            if StartConcorrente < dtIni:
                StartConcorrente = dtIni
            if FinishConcorrente > dtFim:
                FinishConcorrente = dtFim
            # avaliar tempo gasto na concorrência (do processo concorrente)
            tempoConcorrente = td.Timedelta(FinishConcorrente - StartConcorrente).total.minutes
            # avaliar tempo gasto na concorrência (do processo principal)            
            tempoPrincipal = td.Timedelta(dtFim - dtIni).total.minutes
            # avaliar percentual do tempo com concorrência
            percentualConcorrencia = tempoConcorrente/tempoPrincipal
            return percentualConcorrencia
    return 0
```

```python
# Exemplo chamada verificarConcorrencia_Processo
verificarConcorrencia_Processo("Carga GV", datetime(year=2022, month=1, day=3, hour=22, minute=30, second=0),
datetime(year=2022, month=1, day=4, hour=7, minute=24, second=0))
```

```python
# Itera nos processos verificando as concorrências
# Necessário efetuar correção para não verificar concorrência no próprio processo
# Multiplicação por 1 é para converter FALSE e TRUE em zero e um respectivamente
for p in range(len(processos)):
    print(processos[p])
    for i in df[df["Task"]==processos[p]].index: 
         df["Dummy_Carga_GV"][i]                   =  verificarConcorrencia_Processo("Carga GV", df["Start"][i], df["Finish"][i]) *1
         df["Dummy_Fechamento_Contabil"][i]        =  verificarConcorrencia_Processo("Fechamento Contábil", df["Start"][i], df["Finish"][i]) *1        
         df["Dummy_Fechamento_Comissoes"][i]       =  verificarConcorrencia_Processo("Fechamento Comissões", df["Start"][i], df["Finish"][i]) *1
         df["Dummy_Lançamento_SPC"][i]             =  verificarConcorrencia_Processo("Lançamento SPC", df["Start"][i], df["Finish"][i]) *1
         df["Dummy_Lançamento_Serasa"][i]          =  verificarConcorrencia_Processo("Lançamento Serasa", df["Start"][i], df["Finish"][i]) *1
         df["Dummy_Remessa_BACEN"][i]              =  verificarConcorrencia_Processo("Remessa BACEN", df["Start"][i], df["Finish"][i]) *1 
         df["Dummy_Relatorio_Semanal_Contabil"][i] =  verificarConcorrencia_Processo("Relatório Semanal Contábil", df["Start"][i], df["Finish"][i]) *1
         df["Dummy_Fechamento_Contabil_Mensal"][i] =  verificarConcorrencia_Processo("Fechamento Contábil Mensal", df["Start"][i], df["Finish"][i]) *1
         df["Dummy_Lancamento_Cobradoras"][i]      =  verificarConcorrencia_Processo("Lançamento Cobradoras", df["Start"][i], df["Finish"][i]) *1        
    
```
